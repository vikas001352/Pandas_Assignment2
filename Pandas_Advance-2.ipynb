```python
ANS-1



import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]

df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# Access and print the data present in the second row (index 1)
second_row_data = df.iloc[1]
print(second_row_data)

OUTPUT
course_name    Machine Learning
duration                      3
Name: 1, dtype: object


ANS-2



import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]

df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})


Using loc

# Access and print the data at row with label 1 and column with label 'duration'
print(df.loc[1, 'duration'])  # Output: 3

# Access and print the data for all rows and the 'course_name' column
print(df.loc[:, 'course_name'])
# Output:
# 0        Data Science
# 1    Machine Learning
# 2             Big Data
# 3       Data Engineer
# Name: course_name, dtype: object



Using iloc:
    
    
    # Access and print the data at row with index 1 and column with index 1
print(df.iloc[1, 1])  # Output: 3

# Access and print the data for all rows and the 'course_name' column
print(df.iloc[:, 0])
# Output:
# 0        Data Science
# 1    Machine Learning
# 2             Big Data
# 3       Data Engineer
# Name: course_name, dtype: object




ANS-3


import pandas as pd
import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1, 2, 3, 4, 5, 6]

# Creating a dataframe
df1 = pd.DataFrame(np.random.rand(6, 6), columns=columns, index=indices)

# Define the new index
reindex = [3, 0, 1, 2]

# Reindex the DataFrame and store it in 'new_df'
new_df = df1.reindex(reindex)

# Output for new_df.loc[2]
print("Output for new_df.loc[2]:")
print(new_df.loc[2])

# Output for new_df.iloc[2]
print("\nOutput for new_df.iloc[2]:")
print(new_df.iloc[2])


Output for new_df.loc[2]:
column_1    0.523544
column_2    0.566509
column_3    0.030618
column_4    0.791463
column_5    0.954219
column_6    0.249192
Name: 1, dtype: float64

Output for new_df.iloc[2]:
column_1    0.348383
column_2    0.551225
column_3    0.756272
column_4    0.836515
column_5    0.346861
column_6    0.943651
Name: 3, dtype: float64






ANS-4





import pandas as pd
import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1, 2, 3, 4, 5, 6]

# Creating a dataframe
df1 = pd.DataFrame(np.random.rand(6, 6), columns=columns, index=indices)

# (i) Mean of each and every column
mean_values = df1.mean()
print("Mean of each column:")
print(mean_values)

# (ii) Standard deviation of column 'column_2'
std_column_2 = df1['column_2'].std()
print("\nStandard deviation of column 'column_2':", std_column_2)




Mean of each column:
column_1    0.429097
column_2    0.398063
column_3    0.500934
column_4    0.453558
column_5    0.442383
column_6    0.523303
dtype: float64

Standard deviation of column 'column_2': 0.33404253196337056


ANS-5


import pandas as pd
import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1, 2, 3, 4, 5, 6]

# Creating a dataframe
df1 = pd.DataFrame(np.random.rand(6, 6), columns=columns, index=indices)

# Replace data in the second row of column 'column_2' with a string variable
df1.loc[2, 'column_2'] = 'string_data'

# Try to find the mean of column 'column_2'
mean_column_2 = df1['column_2'].mean()
print("Mean of column 'column_2':", mean_column_2)


TypeError: could not convert string to float: 'string_data'

# Replace data in the second row of column 'column_2' with a string variable
df1.loc[2, 'column_2'] = 'string_data'

# Convert 'column_2' back to numeric data type (float)
df1['column_2'] = pd.to_numeric(df1['column_2'], errors='coerce')

# Find the mean of column 'column_2'
mean_column_2 = df1['column_2'].mean()
print("Mean of column 'column_2':", mean_column_2)



Mean of column 'column_2': 0.46852804913542317




ANS-6




Window Specification: This defines the scope of the window, such as the number of data points included or the time period to consider. For example, you can specify a window of size 3 for a moving average or a window that covers the past 7 days for a rolling sum.

Window Partitioning: This divides the data into groups, allowing window functions to be applied separately within each group. For example, you can partition data by a specific column, such as 'category', and calculate statistics within each category.

Some commonly used window functions in pandas include:

rolling(): This creates a rolling window object, which allows you to calculate statistics on a moving window of data. It is often used with functions like mean(), sum(), std(), etc.

expanding(): This creates an expanding window object, which includes all data points up to the current point. It calculates the cumulative statistics as more data points are added.

ewm(): This creates an exponentially weighted moving window object. It is often used for calculating exponentially weighted statistics like exponential moving averages.

Example of using the rolling() function to calculate the moving averag



import pandas as pd

# Sample data
data = {'values': [10, 20, 30, 40, 50]}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate the moving average with a window of size 3
df['moving_average'] = df['values'].rolling(window=3).mean()

print(df)



   values  moving_average
0      10             NaN
1      20             NaN
2      30       20.000000
3      40       30.000000
4      50       40.000000



ANS-7



import pandas as pd

# Get the current timestamp
current_timestamp = pd.Timestamp.now()

# Extract the month and year from the timestamp
current_month = current_timestamp.month
current_year = current_timestamp.year

# Print the current month and year
print("Current Month:", current_month)
print("Current Year:", current_year)

OUTPUT
Current Month: 7
Current Year: 2023



ANS-8



import pandas as pd

def calculate_time_difference(date1, date2):
    # Convert the input strings to pandas datetime objects
    datetime1 = pd.to_datetime(date1)
    datetime2 = pd.to_datetime(date2)

    # Calculate the time difference using timedelta
    time_difference = datetime2 - datetime1

    # Extract days, hours, and minutes from the time difference
    days = time_difference.days
    hours = time_difference.seconds // 3600
    minutes = (time_difference.seconds // 60) % 60

    return days, hours, minutes

def main():
    # Prompt the user to enter the two dates
    date1 = input("Enter the first date (YYYY-MM-DD): ")
    date2 = input("Enter the second date (YYYY-MM-DD): ")

    # Calculate the time difference
    days, hours, minutes = calculate_time_difference(date1, date2)

    # Display the result
    print(f"\nTime Difference: {days} days, {hours} hours, and {minutes} minutes.")

if __name__ == "__main__":
    main()

    
    
    ANS-9
    
    
    import pandas as pd

def convert_column_to_categorical(df, column_name, category_order):
    # Convert the specified column to categorical data type with the specified category order
    df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

def main():
    # Prompt the user to enter the file path
    file_path = input("Enter the file path of the CSV file: ")

    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Prompt the user to enter the column name
    column_name = input("Enter the name of the column to convert to categorical: ")

    # Prompt the user to enter the category order as a comma-separated string
    category_order_str = input("Enter the category order for the column (comma-separated): ")
    category_order = [cat.strip() for cat in category_order_str.split(',')]

    # Convert the specified column to categorical
    convert_column_to_categorical(df, column_name, category_order)

    # Sort the DataFrame based on the specified column
    sorted_df = df.sort_values(by=column_name)

    # Display the sorted data
    print("\nSorted Data:")
    print(sorted_df)

if __name__ == "__main__":
    main()

    
    ANS-10
    
    
    
    Make sure you have the pandas and matplotlib libraries installed. You can install them using pip if you don't have them already:
    
    
    pip install pandas matplotlib

    
    import pandas as pd
import matplotlib.pyplot as plt

def plot_stacked_bar_chart(df):
    # Group the data by product category and calculate the total sales for each category over time
    sales_by_category = df.groupby(['Category', 'Date'])['Sales'].sum().unstack()

    # Plot the stacked bar chart
    sales_by_category.plot(kind='bar', stacked=True, figsize=(10, 6))
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.title('Sales of Each Product Category Over Time')
    plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

def main():
    # Prompt the user to enter the file path of the CSV file
    file_path = input("Enter the file path of the CSV file: ")

    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Convert the 'Date' column to datetime data type
    df['Date'] = pd.to_datetime(df['Date'])

    # Call the function to plot the stacked bar chart
    plot_stacked_bar_chart(df)

if __name__ == "__main__":
    main()

    
    ANS-11
    
    
    
    import pandas as pd
from tabulate import tabulate

def calculate_statistics(df):
    # Calculate the mean, median, and mode of the test scores
    mean_score = df['Test Score'].mean()
    median_score = df['Test Score'].median()
    mode_scores = df['Test Score'].mode()

    # If there are multiple modes, join them as a comma-separated string
    mode_str = ', '.join(map(str, mode_scores))

    return mean_score, median_score, mode_str

def main():
    # Prompt the user to enter the file path of the CSV file containing student data
    file_path = input("Enter the file path of the CSV file containing the student data: ")

    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Calculate the mean, median, and mode of the test scores
    mean_score, median_score, mode_str = calculate_statistics(df)

    # Display the results in a table
    table = [
        ['Statistic', 'Value'],
        ['Mean', mean_score],
        ['Median', median_score],
        ['Mode', mode_str]
    ]

    print(tabulate(table, headers='firstrow', tablefmt='grid'))

if __name__ == "__main__":
    main()

    
    
    
    Enter the file path of the CSV file containing the student data: student_data.csv
+-----------+--------+
| Statistic | Value  |
+-----------+--------+
| Mean      | 83.0   |
| Median    | 85.0   |
| Mode      | 85, 90 |
+-----------+--------+










```
